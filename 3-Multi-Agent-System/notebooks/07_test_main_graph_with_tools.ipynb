```python
# %%
from pprint import pprint

def run_mcq_through_graph(
    question_number: int,
    question: str,
    options_text: str,
    max_revisions: int = 2,
):

    
    initial_state = {
        "question_number": question_number,
        "question": question,
        "options_text": options_text,
        "max_revisions": max_revisions,
        "use_option_verifier": True,
        "use_retriever_tool": True,
    }
    
    print("="*70)
    print("="*70)
    print("سوال:")
    print(question)
    print("\nگزینه‌ها:")
    print(options_text)
    print("\n⏳ در حال اجرای گراف...")
    
    result = graph.invoke(initial_state)

    print("\n✅ اجرای گراف تمام شد.")
    
    final_toon = result.get("final_toon") or {}
    draft_toon = result.get("draft_toon") or {}
    critic_toon = result.get("critic_toon") or {}
    
    print("\n--- نتیجه نهایی (final_toon) ---")
    pprint(final_toon)
    
    print("\n--- آخرین پیش‌نویس (draft_toon) ---")
    pprint(draft_toon)
    
    print("\n--- نظر منتقد (critic_toon) ---")
    pprint(critic_toon)
    
    print("\n--- اطلاعات کمکی ---")
    print("revision_count:", result.get("revision_count"))
    print("max_revisions:", result.get("max_revisions"))
    
    return result
```

```python
# %%
q1 = "ضبط اموال یا اشیا از اختیارات کدام مرجع قضایی است؟"


opts1 = """
1) دادگاه 
2) حسب مورد بازپرس یا دادستان یا دادگاه 
3) بازپرس 
4) دادستان 
"""

res1 = run_mcq_through_graph(
    question_number=110,
    question=q1,
    options_text=opts1,
)
```

```python
# %%
q2 = """
کدام مجازات تعزیری زیر در مقام تخفیف قابل تبدیل به مجازات دیگر نیست؟
"""

opts2 = """
1) مصادره اموال
2) محرومیت از حقوق اجتماعی 
3) حبس درجه 3 
4) شلاق درجه 6
"""

res2 = run_mcq_through_graph(
    question_number=2,
    question=q2,
    options_text=opts2,
)
```

```python
# %%
q3 = """
کدام یک از موارد زیر صحیح است؟
"""

opts3 = """
1) با توافق طرفین امکان از بین بردن سبب انفساخ قرارداد وجود ندارد. 
2) به سبب انفساخ عقد جایز نمی تواند ارادی باشد . 
3) سبب انفساخ عقد ممکن است ارادی باشد اما انفساخ در هر حال قهری است. 
4)  با توجه به قهری بودن انفساخ ،عقد، سبب انفساخ عقد نمی تواند ارادی باشد. 
"""

res3 = run_mcq_through_graph(
    question_number=1,
    question=q3,
    options_text=opts3,
)
```

```python
def debug_tools_output(result):
    """نمایش خروجی ابزارها برای دیباگ"""
    print("="*70)
    print("🔍 DEBUG: Tool Results")
    print("="*70)
    
    # چک کردن verifier output
    verifier = result.get("verifier_output")
    if verifier:
        print("\n📊 Option Verifier Output:")
        print(f"Recommended Answer: {verifier.get('recommended_answer')}")
        print(f"Confidence: {verifier.get('confidence')}/5")
        print("\nScores:")
        for score in verifier.get("scores", []):
            print(f"  گزینه {score['option_number']}: {score['support_level']}")
            print(f"    دلیل: {score['reasoning'][:150]}...")
            print()
    else:
        print("⚠️ verifier_output خالی است!")
    
    # چک کردن tool_results
    tool_results = result.get("tool_results", {})
    if tool_results:
        print("\n🔧 Tool Results Keys:", list(tool_results.keys()))
    else:
        print("⚠️ tool_results خالی است!")
    
    # چک کردن messages
    messages = result.get("messages", [])
    print(f"\n💬 Messages Count: {len(messages)}")
    if messages:
        print("Last 3 messages:")
        for msg in messages[-3:]:
            role = msg.get("role", "unknown")
            content = msg.get("content", "")[:100]
            print(f"  - {role}: {content}...")


# بعد از هر تست اضافه کنید:
debug_tools_output(res1)
debug_tools_output(res2)
debug_tools_output(res3)
```
